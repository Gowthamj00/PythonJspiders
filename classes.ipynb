```python
# Class Variables:
# Class Variables can be accessed by using the class name and object reference.

class Student:
    name='Gowtham'
    age=23
    loc=None
Student.loc='Hyderabad'
#accessing the class variable
# Using className
print(Student.name,Student.age,Student.loc)
#Using ObjectReference
s1=Student()
print(s1.name,s1.age,s1.loc)

#dict form of class.
print(Student)
print(Student())
print(Student.__dict__)
#dict form of object.
print(s1)
print(s1.__dict__)#__dict__ shows only instance attributes, not the inherited class variables.



# Updating the class varaible:
# Class Variables can be modified/Updated by using the class name and object reference.
# By using class name
Student.name='Krish'
print(Student.name)
# If we use class name then it will change in the main class attribute..

#By using object reference
s1.name='Ajay'
print(s1.name)
s2=Student()
print(s2.name)
# If we use object reference then it will change only in the current object reference.

# Deleting the class Variables..
# Class Variables can be modified/Updated by using the class name only.
#By using class Name
del Student.loc
print(Student.__dict__)
```

```python
# Class methods: Peforming crud opeartions on class variables.
class Mobile:
    os='Andriod'
    sim_slots=2
    display='OLED'

    def __init__(self,m_name,camera,price):
        self.m_name=m_name
        self.camera=camera
        self.price=price

    def __str__(self):
        return f'{self.m_name} : {self.camera} : {self.price}'
    
    @classmethod
    def view_clsv_details(cls):
        return f'{cls.os}--{cls.sim_slots}--{cls.display}'
    
    @classmethod
    def create_clsv(cls,crgtype):
        cls.charger_type=crgtype
    
    @classmethod
    def read_clsv(cls):
        return f'{cls.charger_type}'
    
    @classmethod
    def update_os(cls,new_os):
        cls.os=new_os

    @classmethod
    def delete_simslots(cls):
        del cls.sim_slots

mobile1=Mobile('Realme 6','64mp',16000)
print(mobile1)

print(mobile1.view_clsv_details())

mobile1.create_clsv('C-Type')
print(mobile1.read_clsv())

mobile1.update_os('Andriod 16')
print(mobile1.view_clsv_details())

mobile1.delete_simslots()
#print(mobile1.view_clsv_details())
```

```python
#Create a class Restuarant
#Create 5 class Variables.
#Create 5 object variables
#create a class method to modify anyone of the class variable.
#create a class method to add 2 class varaibles.

class Restuarant:
    Resto_name='Paradise'
    Resto_adrs='Palasa'
    Resto_branches=5
    Resto_owner='Gowtham'
    Resto_tag='Never disappoints U'

    def __init__(self,biryani,starters,drinks,waterbottle,riceitem):
        self.biryani=biryani
        self.starters=starters
        self.drinks=drinks
        self.waterbottle=waterbottle
        self.riceitem=riceitem

    def __str__(self):
        return f'{self.biryani} {self.starters} {self.drinks} {self.waterbottle} {self.riceitem}'
        
    @classmethod
    def display_clsv_details(cls):
        return f'{cls.Resto_name}--{cls.Resto_adrs}--{cls.Resto_branches}--{cls.Resto_owner}--{cls.Resto_tag}'

    @classmethod
    def change_owner_name(cls,nowner):
        cls.Resto_owner=nowner

    @classmethod
    def create_2clsv(cls,famous_for,rating):
        cls.Resto_famousFor=famous_for
        cls.Resto_rating=rating

    @classmethod
    def read_updatedclsv(cls):
        return f'{cls.Resto_famousFor}--{cls.Resto_rating}'
    
order1=Restuarant('Dum Birayni','Chicken Lollipop','Sprite',1,'CurdRice')
print(order1)

print(order1.display_clsv_details())

order1.change_owner_name('UMA Venkat')
print(order1.display_clsv_details())

order1.create_2clsv('Hyderbad Dum ki Biryani','*****')
print(order1.read_updatedclsv())

```

```python
class Bank:
    b_name='SBI'
    branch='Nandigam'
    IFSC_code='SBIN0016958'

    def __init__(self,ac_holder,ac_no,age,pin,balance=0):
        self.ac_holder=ac_holder
        self.ac_no=ac_no
        self.age=age
        self.pin=pin
        self.balance=balance

    def __str__(self):
        return f'{self.ac_holder}->{self.ac_no}'
    
    @classmethod
    def add_branch_manager(cls,manager):
        cls.manager=manager
    
    def holder_details(self):
        print('--------User Info--------')
        pin=int(input('Enter Your Pin:'))
        if pin==self.pin:
            return f'''Welcome to {Bank.b_name} Bank.. Hello {self.ac_holder} with accno of {self.ac_no}
and a balance of {self.balance}. Thank You.....'''
        return f'Enter the valid pin'
    
    def deposit(self):
        print('------Deposit Amount-------')
        pin=int(input('Enter Your Pin:'))
        if pin==self.pin:
            deposit_amount=int(input('Enter deposit amount:'))
            if deposit_amount>0:
                self.balance+=deposit_amount
                return f'''Amount deposited to accno {self.ac_no} with deposit amount of {deposit_amount}
and current balance is {self.balance}'''
            else:
                return 'Enter Valid deposit amount(>0)'
        print('Enter valid pin')
        return self.deposit()  

    def withdrawal(self):
        print('------Withdraw Amount-------')
        pin=int(input('Enter Your Pin:'))
        if pin==self.pin:
            withdrawal_amount=int(input('Enter withdrawal amount:'))
            if withdrawal_amount>self.balance:
                self.balance-=withdrawal_amount
                return f'''Amount withdrawn from accno {self.ac_no} with withdrawal amount of {withdrawal_amount}
and current balance is {self.balance}'''
            else:
                return 'Enter Valid deposit amount(>0)'
        print('Enter valid pin')
        return self.withdrawal()

cus1=Bank('Gowtham Jami','20448771736',23,2025)
print(cus1)
print(cus1.holder_details())
print(cus1.deposit())
print(cus1.withdrawal())
```
